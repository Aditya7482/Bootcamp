```python
# Sample inventory: product name -> quantity
inventory = {"apples": 10, "bananas": 5, "oranges": 8}

def increase_quantity(inventory, product, amount):
    """
    Increase the quantity of a product in the inventory.
    If the product doesn't exist, it gets added.
    """
    if product in inventory:
        inventory[product] += amount
    else:
        inventory[product] = amount # start at 'amount' if new product
    return inventory

# Usage
inventory = increase_quantity(inventory, "bananas", 3) 
print(inventory) # {'apples': 10, 'bananas': 8, 'oranges': 8}

inventory = increase_quantity(inventory, "mangoes", 4) 
print(inventory) # {'apples': 10, 'bananas': 8, 'oranges': 8, 'mangoes': 4}
```

```python
inventory = {"apples": 10, "bananas": 5, "oranges": 8}

def delete_key(inventory, product):
    """Delete a product if it exists. No error if it doesn't."""
    inventory.pop(product, None)
    return inventory

delete_key(inventory, "oranges")
print(inventory) # {'apples': 10, 'bananas': 8}

```
