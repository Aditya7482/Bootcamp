```python
# List of numbers
numbers = [10, 15, 22, 38, 40, 57, 65, 79]
even_count = 0
odd_count = 0
# Count even and odd numbers
for num in numbers:
    if num % 2 == 0:
        even_count += 1
    else:
        odd_count += 1
print("Number of even numbers:", even_count)
print("Number of odd numbers:", odd_count)
```
